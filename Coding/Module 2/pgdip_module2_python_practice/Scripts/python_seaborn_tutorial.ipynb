## Installation and import

```python
# %pip install seaborn

import seaborn as sns

import numpy as np
```

## Loading dataset 

```python
data = sns.load_dataset("iris")
```

## Lineplot

```python
sns.lineplot(data,
             x="sepal_length",
             y="sepal_width")
```

```python
sns.set(style="dark")
fmri = sns.load_dataset("fmri") # functional magnetic resonance imaging included in the library

# Plot the responses for different events and regions
sns.lineplot(fmri, # data argument does not need to be specified explicitly
             x="timepoint",
             y="signal",
             hue="region", # equivalent to color in ggplot
             style="event", # equivalent to linetype in ggplot
             )
```

## Histogram

```python
sns.histplot(data.sepal_length,
kde=True, # kernel density estimate
color="m")
```

## Lmplot (linear regression lines + scatter)

```python
sns.set(style="ticks")

# Loading the dataset
df = sns.load_dataset("anscombe")

# Show the results of a linear regression
sns.lmplot(x="x", y="y", data=df)
```

## Scatter plot (stripplot)

```python
import matplotlib.pyplot as plt

# x axis values 
x =['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu'] 
  
# y axis values 
y =[5, 6.7, 4, 6, 2, 4.9, 1.8] 
  
# plotting strip plot with seaborn 
ax = sns.stripplot(x=x, y=y) 
  
# giving labels to x-axis and y-axis 
ax.set(xlabel ='Days', 
       ylabel ='Amount_spend'
       ) # this replace xlabel and ylabel arguments in pyplot
  
# giving title to the plot 
plt.title('My first graph'); 
  
# function to show plot 
plt.show() 
```

```python
## alternative plot

# use to set style of background of plot 
sns.set(style="whitegrid") 
  
# loading data-set 
iris = sns.load_dataset('iris') 
  
# plotting strip plot with seaborn 
# deciding the attributes of dataset on 
# which plot should be made 
ax = sns.stripplot(x='species', y='sepal_length', hue='species', data=iris) 
  
# giving title to the plot 
plt.title('Graph') 

# setting labels

ax.set(xlabel="Species",
       ylabel="Sepal length")
  
# function to show plot 
plt.show() 
```

```python
## some more setups


df = sns.load_dataset('tips') 
sns.stripplot(df,
                x='day', 
              y='total_bill', 
              jitter=True, 
              hue='smoker',
                dodge=True) 
```

## Swarmplot    

Similar to stripplot, but not allowing dots to overlap

```python


# use to set style of background of plot 
sns.set(style="whitegrid") 
  
# loading data-set 
iris = sns.load_dataset('iris') 
  
# plotting strip plot with seaborn 
# deciding the attributes of dataset on 
# which plot should be made 
ax = sns.swarmplot(x='species', y='sepal_length', data=iris) 
  
# giving title to the plot 
plt.title('Graph') 
  
# function to show plot 
plt.show() 
```

## Barplot

```python

# reading the dataset  
df = sns.load_dataset('tips')  
  
# change the estimator from mean to standard deviation  
sns.barplot(x ='sex', 
            y ='total_bill', 
            data = df,   
            palette ='plasma',
            estimator= 'mean' # specify estimator to be shown (uses string call to base function), eg std
            )

```

## Countplot

```python

# reading the dataset  
df = sns.load_dataset('tips')  
  
sns.countplot(x ='sex', data = df)  
```

## Boxplot

```python

sns.boxplot(x='day', y='total_bill', data=df, hue='smoker') 
```

```python
## Violin plot
```

```python
sns.violinplot(x='day', 
               y='total_bill', 
               data=df, 
               hue='sex', 
                
               split=True # whether to show half violin
               ) 

```
