# Pyplot tutorial (basics)

Following the video at https://www.youtube.com/watch?v=nzKy9GY12yo

## Initial import

```python
import matplotlib.pyplot as plt

%matplotlib inline


```

## Simple plot

```python
x = list(range(0,10))
y = list(range(-10,0))


plt.plot(x,y)
```

```python
a = [0, -100, 25, 67, -323]

b = [0,3,7,3,9]

plt.plot(a,b)
```

## Restrict axes

```python
plt.axis([-50,80,
         2,8])

plt.plot(a,b)
```

## Add labels

```python
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")

plt.plot(a,b)
```

## Custom axis tick labels

```python
plt.axis([-50,80,
         2,8])

plt.xticks((-40,-20,0,20,40,60, 80), ("h","e","l","l","o","o","o"))

plt.plot(a,b)
```

## Change color

```python
plt.plot(a,b, color="red")
```

## Histogram

```python
plt.hist(a)
```

```python
plt.hist(x)

plt.hist(y)
```
